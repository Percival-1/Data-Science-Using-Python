# Functions

```python
# Passing the paramneter
def greet(n = 'Unknown'):
    print('Hello', n)

greet('Shubh')
greet(n = 'Madhu')
# greet()   not going to work
greet()

# Passing the Parameters with print
def sum(a = 0, b = 0):
    print(a , b , a+b)

sum(10 , 5)
sum()
sum(3)
sum(b = 4)

# Return
def sum(a = 0 , b = 0):
    return (a+b)

v = sum(10,6)
print(v)

# Multi Return
def arith(a = 0 , b = 0):
    return (a+b) , (a-b) , (a*b)
s = arith(10 , 7)
print(s)
print(type(s))

def square(lst):
    return[i**2 for i in lst]

def cube(lst):
    return [i**3 for i in lst]

def sum(lst):
    a = square(lst)
    b = cube(lst)

    return [x + y for x , y in zip(a , b)]
o = sum([1 , 2 , 3 , 4 , 5])
print(o)

```

# OOPs

```python
class Person:
    name = 'Ashish'
    age = 25

p1 = Person()
print(p1.name)
print(p1.age)

p1.name = 'Prem'
p1.age = 35

print(p1.name)
print(p1.age)

print()

class Math:

    name = 'Pikachu'

    def greet(self):
        print('Hello')
        print(self.name)

    def factorial(self , n):
        f = 1
        for i in range(1 , n+1):
            f *= i
        return f
    
    def mul(self , lst):
        m = 1
        for i in lst:
            m *= i
        return m
    
    def croos_pro(self , l1 , l2):
        lst = [l1[i]*l2[i] for i in range(len(l1))]
        return lst
m1 = Math()
m1.greet()
print(m1.factorial(5))
print(m1.mul([1 , 2 , 3 , 4 , 5]))
print(m1.croos_pro([1 , 2 , 3 , 4 , 5] , [2 , 3 , 4 , 5 , 6]))
```

```python
class agent:

    def __init__(self , name , type):
        self.name = name
        self.type = type
        self.health = 100
        self.alive = True

    def curr_health(self):
        print('Current Health of' , self.name , 'is' , self.health)
    
    def punched(self):
        self.health -=10

    def shooted(self):
        self.health -= 50

    def is_alive(self):
        if self.health <= 0:
            self.alive = False
        return (self.alive)
    
    def info(self):
        print('Name    : ' , self.name)
        print('Type    : ' , self.type)
        print('Health  : ' , self.health)
        print('Alive   : ' , self.is_alive())

p1 = agent('Pikachu', 'Mouse')
p1.shooted()
p1.shooted()
p1.info()

```
