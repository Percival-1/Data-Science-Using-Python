# Dictionary

```python
dct = {1 : 'Pikachu' , 1 : 'Raichu' , 3 : 'Charizad' , 4 : 'Charmander' , 5 : 'squirtle'}
print(dct)

# accessing individual element

print(dct[1])

print(dct.get(3))

# Addidng and updating key-value pairs

dct[6] = 'Doraemon'

print(dct)

del dct[6]
print(dct)

dct.pop(1)
print(dct)

dct.popitem() #delete the last element
print(dct)
dct.clear()
print(dct)
# traversing in a dictionary

dct = {1 : 'Pikachu' , 7 : 'Raichu' , 3 : 'Charizad' , 4 : 'Charmander' , 5 : 'squirtle'}
print(dct.keys()) # return list of keys
print(dct.values()) # return list of values

for i in dct.keys():
    print(dct[i])

print(dct.items()) # gives list of keys and values in tuple format

#for  i in dct.items():
#    print(i[0], i[1])

for k, v in dct.items():
    print(k , v)

# check if key is present

print(1 in dct)
print('1' in dct) # string 1 is not present only int  1 is present

# updating two Dictionary

dct_1 = {1 : 'A' , 2 : 'B' , 3 : 'C'}
dct_2 = {1 : 'a' , 2 : 'b' , 3 : 'c'}

dct_1.update(dct_2)
print(dct_1)
```

# Multi Dimensional Dictionary

```python
dct = {1 : {'Name' : 'Pikachu' , 'Type' : 'Normal'} ,
       2 : {'Name' : 'Grayninja' , 'Type' : 'Special'} ,
       3 : {'Name' : 'Mewtow' , 'Type' : 'Legendary'}}
print(dct)

# Accessing the elements

print(dct[1])

print(dct[1]['Name'])

for i in dct.keys():
    for j in dct[i].keys():
        print(j  , dct[i][j])

# Addidng , updating and deleting

dct[4] = {'Name' : 'Blastoise' , 'Type' : 'Normal'}

dct[3]['Name'] = 'Balbasour'
print(dct)

del dct[3]
print(dct)

# Going a level deeper with Performance

dct = {1 : {'Name' : 'Pikachu'   , 'Type' : 'Normal'    , 'Performance' : {'Attack' : 'A' , 'Defence' : 'B' , 'Agility' : 'S'}} ,
       2 : {'Name' : 'Grayninja' , 'Type' : 'Special'   , 'Performance' : {'Attack' : 'A' , 'Defence' : 'A' , 'Agility' : 'S'}} ,
       3 : {'Name' : 'Mewtow'    , 'Type' : 'Legendary' , 'Performance' : {'Attack' : 'S' , 'Defence' : 'S' , 'Agility' : 'S'}}}

print(dct)

for i in dct:
    print(i, dct[i]['Name'], dct[i]['Performance'])

```

# Dictionary Comprehension

```python
dct = {x : x**2 for x in range(1 , 11)}
print(dct)

# Dictionaries with Conditions
dct = {x : x**2 for x in range(1 , 11) if(x%2 == 0)}
print(dct)

# dictionaries from list
lst = ['Apple' , 'Banana'  ,'Grapes']
dct = {items : len(items) for items in lst}
print(dct)
dct = {(x+1) : lst[x] for x in range(len(lst))}
print(dct)

# Special case with String

dct = {('num_'+ str(i)) : i for i in range(1 , 11)}
print(dct)

dct = {k : v for k , v in dct.items() if v%3 == 0}
print(dct)

l_1 = ['a' , 'b' , 'c' , 'd']
l_2 = [1 , 2 , 3 , 4]

dct = { l_2[x] : l_1[x] for x in range(len(l_1))}
print(dct)

# Matrix
matrix = [[1 , 2 , 3] , [4 , 5 , 6] , [7 , 8 , 9]]
dct = {(i+1 , j+1) : matrix[i][j] for i in range(len(matrix)) for j in range(len(matrix[i]))}
print(dct)
```

# Sets

```python
st ={1 , 2 , 3 , 4 , 5}
print(st)
print(type(st))

# Adding
st.add(6)
print(st)

# Pop
p = st.pop()
print(st)

# Removing
st.discard(3)
print(st)

# Iterating
for i in st:
    print(i)

# checking the value
print(2 in st)
print(3 in st)

# set operations

st1 = {1 , 2 , 3 , 4 , 5}
st2 = {3 , 5 , 6 , 7 , 8}
print('Union' , st1 | st2)
print('Intersection' , st1 & st2)
print('Differnce' , st1 - st2)
print('Difference' , st2 - st1)
print('Symmetric Diff' , st1^st2)

# Clearing
st.clear()
print(st)

# tuple
```
